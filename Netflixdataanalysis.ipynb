Q2-> In which year highest number of the Tv Shows  and Movies were released? Show with Bar Graph

```python
#first we see the datatype of our columns
data.dtypes
```

```python
import pandas as pd
```

```python
# changing the datatype
data['Date_N'] = pd.to_datetime(data['Release_Date'], errors='coerce')
```

```python
data.dtypes
```

```python
data['Date_N'].dt.year.value_counts()
```

```python
#creatig Bar graph
import matplotlib.pyplot as plt
plt.figure(figsize= (10,5))
ax =sns.barplot(data['Date_N'].dt.year.value_counts(),palette="pastel")
ax.bar_label(ax.containers[0],label_type='edge',fontsize=10, padding=3)
plt.title("TV shows release count yearly",fontsize=14, fontweight="bold")
plt.xlabel("Date_N", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout() 
plt.show()
```

Q3->How many Movies and Tv shows are in the dataset?Show with Bar Graph


```python
data['Category Count'] = data.groupby('Category')['Category'].transform('count')
```

```python
colors = sns.color_palette("pastel")

# Create the count plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=data, x='Category', order=data['Category'].value_counts().index, palette=colors
)

# Add bar labels
ax.bar_label(ax.containers[0])

# Add a title to the plot
plt.title("Different Types of Shows and Their Count", fontsize=14, fontweight="bold")

# Show the plot
plt.show()
```
