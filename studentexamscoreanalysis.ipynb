Gender Distribution

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df,x = "Gender", palette="pastel")
ax.bar_label(ax.containers[0],label_type='edge',fontsize=10, padding=3)
plt.title("Gender Distribution ",fontsize=14, fontweight="bold")
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout() 
plt.show()
```

#from above chart we have analysed that :
# number of females  data is more than the numbers of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot = True, cmap="coolwarm", linewidths=0.5,linecolor="white", )
plt.title("Relatio btw student score and parents education",fontsize=14, fontweight="bold")

plt.tight_layout() 
plt.show()
```

from the above chat we have conculded that the education of parents have great impact

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True, linewidths=0.5,linecolor="white", )
plt.title("Relation btw student score and parent marital status",fontsize=14, fontweight="bold")
plt.show()
```

From above chart we have concluded that there is no impact on student scores due to their parents marital status

Distribution of ethnic groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()
l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
color = sns.color_palette("spring")
plt.pie(mlist,labels = l, autopct= "%1.2f%%",colors= color,shadow=True)

plt.title("Distribution of Ethnic Group",fontsize=14, fontweight="bold")
plt.show()

```
